```python
# remove insignificant variables => Months since last claim, number of open complaints, months since last policy inception
# number of policies
Xsig = X.drop(['months_since_last_claim', 'number_of_open_complaints','months_since_policy_inception','number_of_policies'], axis=1)
Xsig
```

```python
# make it look like a more normal distribution

def boxcox_transform(Xsig):
    numeric_cols = Xsig.select_dtypes(np.number).columns
    _ci = {column: None for column in numeric_cols}
    for column in numeric_cols:
        # since i know any columns should take negative numbers, to avoid -inf in df
        Xsig[column] = np.where(Xsig[column]<=0, np.NAN, Xsig[column]) 
        Xsig[column] = Xsig[column].fillna(Xsig[column].mean())
        transformed_data, ci = stats.boxcox(Xsig[column])
        Xsig[column] = transformed_data
        _ci[column] = [ci] 
    return Xsig, _ci
Xsig, _ci = boxcox_transform(Xsig)
Xsig.head()
```

```python
# See the changes in the distributions
for column in Xsig.columns:
    sns.displot(Xsig[column])
    plt.show()
```
